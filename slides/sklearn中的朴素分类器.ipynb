通过csv文件存放数据集到本地，通过panda.read_csv将数据导入为python变量


```python
import pandas as pd
xigua_df = pd.read_csv('./data/xigua.csv')
```

```python
xigua_df
```

对数据进行一个洗牌操作，模拟现实中随机收集西瓜的情况：

```python
from sklearn.utils import shuffle
xigua_df = shuffle(xigua_df, random_state=2022)
xigua_df
```

```python
xigua_df.columns
```

我们先专注在数据值的特征上：

```python
X=xigua_df[['fea_hantangdu', 'fea_midu']]
y=xigua_df['label_haogua']
```

```python
X
```

<div style="color:white;display:fill;border-radius:8px;
            background-color:#323232;font-size:150%;
            font-family:Nexa;letter-spacing:0.5px">
    <p style="padding: 8px;color:white;"><b>切分为训练数据与测试数据</b></p>
</div>

简单将最后2个样本作为测试数据


```python
te_num = 2
X_tr = X[:len(X)-te_num]
y_tr = y[:len(X)-te_num]
X_te =  X[len(X)-te_num:]
y_te = y[len(X)-te_num:]
```

```python
len(X_tr), len(X_te)
```

<div style="color:white;display:fill;border-radius:8px;
            background-color:#323232;font-size:150%;
            font-family:Nexa;letter-spacing:0.5px">
    <p style="padding: 8px;color:white;"><b>使用连续的特征进行预测</b></p>
</div>

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_tr, y_tr)
```

```python
y_pred = model.predict(X_te)
y_pred
```

```python
score = model.score(X_te, y_te)
print('Accuracy Score: ', score)
```

<div style="color:white;display:fill;border-radius:8px;
            background-color:#323232;font-size:150%;
            font-family:Nexa;letter-spacing:0.5px">
    <p style="padding: 8px;color:white;"><b>使用离散的特征进行预测</b></p>
</div>

分隔出来离散型特征的数据

```python
discrete_cols = ['fea_seze', 'fea_gendi', 'fea_qiaosheng', 'fea_wenli', 'fea_qibu','fea_chugan']

X_ =xigua_df[discrete_cols]
X_
```

调用sklearn里的编码工具进行处理

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
X_ = enc.fit_transform(X_)
X_
```

重新做预测

```python
te_num = 2
X_tr = X_[:len(X)-te_num]
y_tr = y[:len(X)-te_num]
X_te =  X_[len(X)-te_num:]
y_te = y[len(X)-te_num:]

from sklearn.naive_bayes import CategoricalNB
model = CategoricalNB()
model.fit(X_tr, y_tr)
```

```python
model.predict(X_te)
```

```python
score = model.score(X_te, y_te)
print('Accuracy Score: ', score)
```

```python
y_pred = model.predict(X_te)
y_pred
```
